### To multiply 2 matrices the rows of one would have to be equal to the column of one

```python
import numpy as np

np.random.seed(42)

m = 100
X = 2 * np.random.rand(m,1)
y = 4 + 3 * X + np.random.randn(m,1)
```

```python
print(X)
```

```python
y.shape
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (20,9))
ax.scatter(X, y)

```

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X)
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
theta_best
```

```python
X_new_b = add_dummy_feature(np.array([[0],[2]]))
X_new = np.array([[0],[2]])
y_predict = X_new_b @ theta_best
y_predict
```

```python
fig, ax = plt.subplots(figsize = (20,9))
ax.scatter(X, y, color = 'blue', label = 'Data')
ax.plot(X_new, y_predict, color = 'red', label = 'Prediction')
ax.legend()
plt.show()
```
